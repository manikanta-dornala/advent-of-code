```python
import numpy as np
import pandas as pd
import warnings
from functools import lru_cache
warnings.filterwarnings('ignore')
```

```python
data = []
with open('../inputs/2024/Day4.txt', 'r') as f:
    for row in f.readlines():
        data.append(row.replace(' ','').replace('\n', ''))
data

# data = [
#     'MMMSXXMASM',
#     'MSAMXMSMSA',
#     'AMXSXMAAMM',
#     'MSAMASMSMX',
#     'XMASAMXAMM',
#     'XXAMMXXAMA',
#     'SMSMSASXSS',
#     'SAXAMASAAA',
#     'MAMMMXMMMM',
#     'MXMXAXMASX',
# ]
# [print(x) for x in data]

height = len(data)
width = len(data[0])
data = [row for row in data if len(row)==width]
height = len(data)
width = len(data[0])
check = 'XMAS'
count = 0
mask = []
for y in range(height):
    mask.append([
        '.' for x in range(width)]
    )

for y in range(height):
    for x in range(width):
        if data[y][x] != 'X':
            continue

        # +x 
        if x+3 < width:
            piece = data[y][x] + data[y][x+1] + data[y][x+2] + data[y][x+3]
            if piece == check:
                count += 1
                # print('+x', x, y)
                mask[y][x], mask[y][x+1], mask[y][x+2], mask[y][x+3] = data[y][x], data[y][x+1], data[y][x+2], data[y][x+3]
        # -x
        if x-3 >= 0:
            piece = data[y][x] + data[y][x-1] + data[y][x-2] + data[y][x-3]
            if piece == check:
                count += 1
                # print('-x', x, y)
                mask[y][x], mask[y][x-1], mask[y][x-2], mask[y][x-3] = data[y][x], data[y][x-1], data[y][x-2], data[y][x-3]
        
        # +y
        if y+3 < height:
            piece = data[y][x] + data[y+1][x] + data[y+2][x] + data[y+3][x]
            if piece == check:
                count += 1
                # print('+y', x, y)
                mask[y][x], mask[y+1][x], mask[y+2][x], mask[y+3][x] = data[y][x], data[y+1][x], data[y+2][x], data[y+3][x]
        # -y
        if y-3 >= 0:
            piece = data[y][x] + data[y-1][x] + data[y-2][x] + data[y-3][x]
            if piece == check:
                count += 1
                # print('-y', x, y)
                mask[y][x], mask[y-1][x], mask[y-2][x], mask[y-3][x] = data[y][x], data[y-1][x], data[y-2][x], data[y-3][x]

        # +x+y
        if x+3 < width and y+3 < height:
            piece = data[y][x] + data[y+1][x+1] + data[y+2][x+2] + data[y+3][x+3]
            if piece == check:
                count += 1
                # print('+x+y', x, y)
                mask[y][x], mask[y+1][x+1], mask[y+2][x+2], mask[y+3][x+3] = data[y][x], data[y+1][x+1], data[y+2][x+2], data[y+3][x+3]
        
        # +x-y
        if x+3 < width and y-3 >= 0:
            piece = data[y][x] + data[y-1][x+1] + data[y-2][x+2] + data[y-3][x+3]
            if piece == check:
                count += 1
                # print('+x-y', x, y)
                mask[y][x], mask[y-1][x+1], mask[y-2][x+2], mask[y-3][x+3] = data[y][x], data[y-1][x+1], data[y-2][x+2], data[y-3][x+3]

        # -x-y
        if x-3 >= 0 and y-3 >= 0:
            piece = data[y][x] + data[y-1][x-1] + data[y-2][x-2] + data[y-3][x-3]
            if piece == check:
                count += 1
                # print('-x-y', x, y)
                mask[y][x], mask[y-1][x-1], mask[y-2][x-2], mask[y-3][x-3] = data[y][x], data[y-1][x-1], data[y-2][x-2], data[y-3][x-3]

        # -x+y
        if x-3 >= 0 and y+3 < height:
            piece = data[y][x] + data[y+1][x-1] + data[y+2][x-2] + data[y+3][x-3]
            if piece == check:
                count += 1
                # print('-x+y', x, y)
                mask[y][x], mask[y+1][x-1], mask[y+2][x-2], mask[y+3][x-3] = data[y][x], data[y+1][x-1], data[y+2][x-2], data[y+3][x-3]

print(count)
```

```python
data = []
with open('../inputs/2024/Day4.txt', 'r') as f:
    for row in f.readlines():
        data.append(row.replace(' ','').replace('\n', ''))
data

# data = [
#     'MMMSXXMASM',
#     'MSAMXMSMSA',
#     'AMXSXMAAMM',
#     'MSAMASMSMX',
#     'XMASAMXAMM',
#     'XXAMMXXAMA',
#     'SMSMSASXSS',
#     'SAXAMASAAA',
#     'MAMMMXMMMM',
#     'MXMXAXMASX',
# ]
# [print(x) for x in data]

height = len(data)
width = len(data[0])
data = [row for row in data if len(row)==width]
height = len(data)
width = len(data[0])
check = 'MAS'
count = 0
mask = []
for y in range(height):
    mask.append([
        '.' for x in range(width)]
    )

for y in range(1,height-1):
    for x in range(1,width-1):
        if data[y][x] != 'A':
            continue

        diag_1: str = data[y-1][x-1] + data[y][x] + data[y+1][x+1]
        diag_2: str = data[y+1][x-1] + data[y][x] + data[y-1][x+1]

        if (diag_1 == check or diag_1[::-1] == check) and (diag_2 == check or diag_2[::-1] == check):
            count += 1
print(count)
```
